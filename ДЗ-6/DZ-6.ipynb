```python
import pandas as pd
 
 
def load_data(file_path, nrows=15):
    return pd.read_csv(file_path, nrows=nrows)
 
 
def categorize_age(age):
    if age > 25:
        return "Старше 25"
    else:
        return "Младше или равен 25"
 
def process_age_column(df):
    df['age_category'] = df["age"].apply(categorize_age)
 
    return df
 
# df - data frame
 
def filter_data_by_conditions(df):
    filtered_df = df[(df["sex"] == "male") & (df["age"] > 30)]
    return filtered_df
 
def print_age_categories(df):
    for idx, category in enumerate(df['age_category']):
        print(f"{idx}, {category}")
 
def main():
    print("Первое задание")
 
    df = load_data("web_clients_correct_доп материал к тема 6, 7.csv")
    df = process_age_column(df)
    print_age_categories(df)
 
    print("Второе задание")
    filtered_df = filter_data_by_conditions(df)
    print(filtered_df)
main()
```
