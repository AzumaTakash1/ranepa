```python
while True:
    number = int(input("Введите 6 значное число: "))
 
    num_str = str(number)
    if len(num_str) == 6:
        first_sum = sum(int(digit) for digit in num_str[:3])
        last_sum = sum(int(digit) for digit in num_str[3:])
 
        if first_sum == last_sum:
            print("Счастливый билет")
        else:
            print("Несчастливый билет")
    else:
        print("Не 6 значное число")
```
