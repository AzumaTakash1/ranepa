```python
while True:
    year = int(input("Введите 4х значный год: "))
    if year < 1000 or year > 9999:
        print("Неправильное число")
        continue
    if (year % 4) == 0 and (year % 100 != 0 or year % 400 == 0):
        print(year, "високосный год")
    else:
        print(year, "обычый год")
```
