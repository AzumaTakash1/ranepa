```python
while True:
    input_boys = input("Введите мальчиков через пробел:")
    input_girls = input("Введите девочек через пробел:")
 
    boys_list = input_boys.split(" ")
    girls_list = input_girls.split(" ")
    if len(boys_list) != len(girls_list):
        print("Внимание, ктото может остаться без пары")
        if len(boys_list) < len(girls_list):
            print(girls_list[-1])
        else:
            print(boys_list[-1])
        input_decision = input("Хотите ли продолжить(y/n):")
        if input_decision == "y":
            continue
        else:
            break
 
    else:
        sorted_boys = sorted(boys_list)
        sorted_girls = sorted(girls_list)
        print("Идеальные пары:")
        for i in range(len(sorted_boys)):
            print(f"{sorted_boys[i]} и {sorted_girls[i]}")
```
