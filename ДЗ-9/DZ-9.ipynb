```python
import random
 
def play_once() -> float:
    payouts = {
        777: 200,
        999: 100,
        555: 50,
        333: 15,
        111: 10,
    }
 
    number = random.randint(0, 999)
 
    win = payouts.get(number, 0)
    net_gain = win - 1
 
    return net_gain
 
def simulate_games(n_games: int) -> float:
    total = 0
 
    for _ in range(n_games):
        total += play_once()
 
    average_gain = total / n_games
    return average_gain
 
def theoretical_expectation() -> float:
    payouts = {
        777: 200,
        999: 100,
        555: 50,
        333: 15,
        111: 10,
    }
 
    expected_win = sum(payouts.values()) / 1000
 
    expected_net = expected_win - 1
    return expected_net
 
def main():
    n_games = 1_000_0000
    avg_gain_sim = simulate_games(n_games)
    print(f"Результат моделирования ({n_games} игр): {avg_gain_sim:.4f} руб за игру")
 
    avg_game_theor = theoretical_expectation()
    print(f"Теоритичесткое мат. ожидание: {avg_game_theor:.4f} руб. за игру")
 
 
if __name__ == "__main__":
    main()
```
